## Cropping



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('images/girl.jpg')
plt.imshow(image[...,::-1])
```

```python
height, width = image.shape[:2]
print(height, width)
cropped = image[200:600, 250:630]
plt.imshow(cropped[...,::-1])
```

```python

height, width = image.shape[:2]

# Simply use indexing to crop out the rectangle we desire
cropped = image[height//3:4*height//5 , width//4:4*width//5]

plt.imshow(cropped[...,::-1])
```
